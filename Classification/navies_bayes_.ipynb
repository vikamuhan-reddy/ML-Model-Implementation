```python
import pandas as pd
df = pd.read_csv("C:/Users/Shoba/Downloads/titanic (1).csv")
df
```

```python
df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1)
df
```

```python
target = df.Survived
target
```

```python
x = df.drop(['Survived'],axis=1)
x
```

```python
dum = pd.get_dummies(x.Sex)
dum = dum.replace({False:0,True:1})
dum
```

```python
x = pd.concat([x,dum],axis=1)
x = x.drop(['Sex'],axis=1)
x
```

```python
x['Age']= x['Age'].fillna(x['Age'].mean())
x
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test= tts(x,target,test_size=0.2)
```

```python
# navie bayes model
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
