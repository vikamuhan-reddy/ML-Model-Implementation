```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# ignore error
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("C:/Users/Shoba/Downloads/Melbourne_housing_FULL.csv")
data
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
```

```python
data = data[cols_to_use]
```

```python
data.isna().sum()
```

```python
col_to_zero  = ['Distance','Propertycount','Bedroom2','Bathroom','Car']
```

```python
data[col_to_zero] = data[col_to_zero].fillna(0)
data
```

```python
data.isna().sum()
```

```python
data['Landsize'] = data['Landsize'].fillna(np.mean(data['Landsize']))
data['BuildingArea'] = data['BuildingArea'].fillna(np.mean(data['BuildingArea']))
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
data
```

```python
data.isna().sum()
```

```python
data = pd.get_dummies(data,drop_first=True)
data
```

```python
data= data.replace({False:0,True:1})
data
```

```python
x = data.drop(['Price'],axis=1)
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression as LR
model = LR()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
l_model = Lasso(alpha=100,max_iter=14,tol=0.1)
```

```python
l_model.fit(x_train,y_train)
```

```python
l_model.score(x_test,y_test)
```

```python
r_model = Ridge(alpha=50,)
```

```python
r_model.fit(x_train,y_train)
```

```python
r_model.score(x_test,y_test)
```
